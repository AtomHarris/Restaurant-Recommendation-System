```python
# importing libraries
import pandas as pd
```

```python
# File Path to JSON file
file_path = 'yelp_academic_dataset_business.json'

# Load the JSON file into a DataFrame,
business = pd.read_json(file_path, lines=True)

# Displaying first 5 entries
business.head()
```

```python
# Filter the DataFrame to keep only rows that contain the word "restaurant"
restaurants_df = business[business['categories'].str.contains("restaurant", case=False, na=False)]

# Display the first few rows of the filtered DataFrame
restaurants_df.head()
```

```python
# Saving restaurant data to csv 
restaurants_df.to_csv('data/restaurants.csv', index=False)
```

```python
import pandas as pd

# Path to the JSON file
json_file_path = 'yelp_academic_dataset_review.json'

# Define a chunk size
chunk_size = 10000

# Initialize an empty list to hold chunks
chunks = []

# Read the JSON file in chunks
with pd.read_json(json_file_path, lines=True, chunksize=chunk_size) as reader:
    for chunk in reader:
        # Process each chunk (e.g., filter, aggregate)
        chunks.append(chunk)

# Concatenate all chunks into a single DataFrame
df = pd.concat(chunks, ignore_index=True)

```
